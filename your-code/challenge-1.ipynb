**0. Load seaborn dataset "tips"**


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

import ssl
ssl._create_default_https_context = ssl._create_unverified_context

tips = pd.read_csv('tips.csv')
tips.head()
```

**1. What is the prefered day for eating?**

```python
sns.countplot(x = tips.day);
```

In the graph we can see that the favourite day for eating is Saturday, followed by Sunday, Thursday and Friday

**2. What days of the week has more men than women payers?**

```python
sns.countplot(x = tips.day,
             hue = tips.sex);
```

Sundays, Saturdays and Fridays
Sunday is the day in which the difference between sexs is higher; on Fridays the difference is almost non-existent

**3. Do you think bill amounts change considerably depending on smoking?**

```python
sns.histplot(x = tips.total_bill,
             hue = tips.smoker);
```

Higher bills seem to be payed by non-smokers, but it is not a strong correlation

**4. Is tip correlated to total_bill? Draw a scatter plot**

```python
sns.scatterplot(x = tips.total_bill,
               y = tips.tip);
```

As well as in the previous plot, there is a correlation between both variables, in which some clientes give more tip when the bill is higher, and a reduce tip when the bill is lower, however the graph doesn't show a strong correlation. There are many cases in with the tip seems to be independent of the bill.
