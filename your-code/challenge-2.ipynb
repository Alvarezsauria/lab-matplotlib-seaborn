```python
sns.boxplot(x = 'Months_encoded',
           y = 'Steps', data = fitbit);
```

It's also remarcable the wide variability of data collected from users on March, May and July

```python
sns.histplot(x = 'Minutes Lightly Active',
              hue = 'Work_or_Weekend',
             data = fitbit)
```
